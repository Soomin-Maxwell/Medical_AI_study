cv2.findContours(image, mode, method)

- mode : Contours들을 찾는 방법 <br/>
RETR_EXTERNAL : 바깥쪽 Line만 찾기<br/>
RETR_LIST : 모든 Line을 찾지만, Hierarchy 구성 X<br/>
RETR_TREE : 모든 Line을 찾으며, 모든 Hierarhchy 구성 O
-  method : Contours 들을 찾는 근사치 방법<br/>
CHAIN_APPROX_NONE : 모든 Contour포인트 저장<br/>
CHAIN_APPROX_SIMPLE : Contour Line을 그릴 수 있는 포인트만 저장 <br/>


----

cv2.drawContours(image, contours, contour_index, color, thickness) : Contour들을 그리는 함수 

- contour_index : 그리고자하는 Conoutrs line

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Image.jpeg")
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(image_gray, 127, 255, 0)
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB))
```

```python
contours,hierarchy = cv2.findContours( thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
image_2 = cv2.drawContours(image, contours, -1, (0, 255, 0), 3)
plt.imshow(cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB))
plt.show() 
```
