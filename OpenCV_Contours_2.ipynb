cv2.boundingRect(contour) : Contour 를 포함하는 사각형을 그림
- 사각형의 X,Y 좌표와 너비, 높이 반환

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image_2.png')

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Gray 이미지로 전환
ret, thresh = cv2.threshold(image_gray, 70, 255, 0) # binary 이미지로 전환
#thresh = cv2.bitwise_not(thresh) # 반전 처리
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB)) # 이미지 확인 
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0,0,255), 15)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # 이미지 확인 
plt.show()

contour = contours[0]
x,y,w,h = cv2.boundingRect(contour)
#print(x,y,w,h)
image_2 = cv2.rectangle(image,(x,y), (x+w, y+h), (0,255,0), 15)
plt.imshow(cv2.cvtColor(image_2,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg')

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Gray 이미지로 전환
ret, thresh = cv2.threshold(image_gray, 200, 255, 0) # binary 이미지로 전환
thresh = cv2.bitwise_not(thresh) # 반전 처리
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB)) # 이미지 확인 
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0,0,255), 15)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # 이미지 확인 
plt.show()

contour = contours[0]

hull = cv2.convexHull(contour)
image = cv2.drawContours(image, [hull], -1, (0,255,0), 15)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg')

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Gray 이미지로 전환
ret, thresh = cv2.threshold(image_gray, 200, 255, 0) # binary 이미지로 전환
thresh = cv2.bitwise_not(thresh) # 반전 처리
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB)) # 이미지 확인 
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0,0,255), 15)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # 이미지 확인 
plt.show()

contour = contours[0]
epsilon = 0.01 * cv2.arcLength(contour, True)
approx = cv2.approxPolyDP (contour,epsilon, True)
image = cv2.drawContours(image, [approx], -1, (0,255,0), 15)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg')

image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # Gray 이미지로 전환
ret, thresh = cv2.threshold(image_gray, 200, 255, 0) # binary 이미지로 전환
thresh = cv2.bitwise_not(thresh) # 반전 처리
plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_GRAY2RGB)) # 이미지 확인 
plt.show()

contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
image = cv2.drawContours(image, contours, -1, (0,0,255), 15)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # 이미지 확인 
plt.show()

contour = contours[0]

area = cv2.contourArea(contour)
print(area)

length = cv2.arcLength(contour, True)
print(length)

M = cv2.moments(contour)
print(M)
```
